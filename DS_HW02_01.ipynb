**Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"**

```python
import pandas as pd
import numpy as np
table_url = "https://uk.m.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
tmp = pd.read_html(table_url, thousands="", decimal = ",",
                   match='Коефіцієнт народжуваності в регіонах України',
                   encoding="utf-8"
                    )
table = pd.DataFrame(tmp[0])
table
```

```python
missed_to_replace = new_table.apply(pd.to_numeric, errors='ignore').mean(axis=0, numeric_only=True).round()
print(missed_to_replace)
table_fillna_mean = new_table.fillna(missed_to_replace).apply(pd.to_numeric, errors='ignore')
table_fillna_mean
```

# New Section

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні


```python
avg = table_fillna_mean.mean(numeric_only=True)['2019'].round()
print(f"Mean in UA 2019: {avg}")

table_fillna_mean[['Регіон', '2019']][table_fillna_mean['2019'] > avg]
```

```python
el = table_fillna_mean.shape[0]
print(el)
avg_UA = table_fillna_mean.loc[el-1]['2019'].round()
print(f"Mean in UA 2019: {avg_UA}")

table_fillna_mean[['Регіон', '2019']][table_fillna_mean['2019'] > avg_UA]
```

У якому регіоні була найвища народжуваність у 2014 році?


```python
table_fillna_mean[['Регіон', '2014']][table_fillna_mean['2014'] == table_fillna_mean['2014'].max()]
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році


```python
import pandas as pd
import matplotlib.pyplot as plt

plt.bar(table_fillna_mean['Регіон'], table_fillna_mean['2019'], color=['y', 'b'])
plt.xlabel('Регіон', fontsize='medium', color='k')
plt.xticks(rotation=90, fontsize='small')
plt.ylabel('Коефіціент народжуваністі', fontsize='medium', color='k')
plt.title('Діаграма народжуваності по регіонам в 2019 році', fontsize='x-large', color='b')
plt.tight_layout()
plt.show()
```

```python
df = table_fillna_mean
fig, axs = plt.subplots(2, 2, constrained_layout=True)
axs[0][0].plot(df['1950'], label='1950')
axs[0][1].plot(df['1960'], label='1960')
axs[1][0].plot(df['2014'], label='2014')
axs[1][1].plot(df['2019'], label='2019')

axs[0][0].set_title('1950')
axs[0][1].set_title('1960')
axs[1][0].set_title('2014')
axs[1][1].set_title('2019')

plt.suptitle("Коефіцієнт народжуваності в Україні.", color='b')
plt.legend(labels=df)
plt.show()
```

```python
df = table_fillna_mean
regions = ["Крим", "Київ", "Україна"]

for region in regions:
    region_data = df[df["Регіон"] == region]
    plt.plot(df.columns[2:], region_data.iloc[0, 2:], label=region)

plt.title("Порівняння народжуваності: Київ vs Крим vs Україна")
plt.xlabel("Рік")
plt.ylabel("Народжуваність")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
